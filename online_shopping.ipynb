Data Cleaning

```python
import re
df['Products'] = df['Các mặt hàng bạn thường lựa chọn:'].str.split(r',(?![^(]*\))')
df = df.explode('Products').reset_index(drop=True)
df.head()
```

Data Cleaning 2

```python
replacement_dict = {
    'Thực phẩm ăn liền (bánh mì, cháo, đồ ăn nhanh, mì miến,…)': 'Thức ăn nhanh (xúc xích, giò chả, lạp xưởng, bánh bao, nem, đậu hũ,…)',
    ' Thực phẩm ăn liền (bánh mì, cháo, đồ ăn nhanh, mì miến,…)' : 'Thức ăn nhanh (xúc xích, giò chả, lạp xưởng, bánh bao, nem, đậu hũ,…)',
    ' Thức ăn nhanh (xúc xích, giò chả, lạp xưởng, bánh bao, nem, đậu hũ,…)' : 'Thức ăn nhanh (xúc xích, giò chả, lạp xưởng, bánh bao, nem, đậu hũ,…)'
}

df['Products'] = df['Products'].replace(replacement_dict)

# Applying the replacements
is_duplicate = df.duplicated(subset=['PID', 'Products'])
df = df[~is_duplicate]
```

```python
# Groupby by the Transaction Ids and count items
basket = df.groupby(
              by=['PID',
                  'Products'])['Products'].count().reset_index(name='Item_Count')

# Pivot table by the transaction and convert item count to boolean
basket = basket.pivot_table(
              index='PID',
              columns='Products',
              values='Item_Count',
              aggfunc='sum').fillna(0).astype(bool)

basket
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

# Apply Apriori algorithm
pd.set_option('display.max_colwidth', 200)

frequent_itemsets = apriori(basket, min_support=0.2, use_colnames=True)
display(frequent_itemsets.sort_values(by='support', ascending=False).head(20))
```

```python
# Compute association rules
rules = association_rules(frequent_itemsets, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head(10)
```
